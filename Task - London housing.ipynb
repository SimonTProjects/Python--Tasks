# 2b. Add 'MONTH' as second coloumn, whihc controls months only

```python
udf2['month'] = pd.DatetimeIndex(udf2["date"]).month
udf2
```

```python
udf2 = udf2.drop(columns=['year', 'month'])
udf2
```
