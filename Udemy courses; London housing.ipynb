```python
udf2 = pd.read_csv(r"C:\Users\simon\Desktop\iTeam - Data Skills Bootcamp\WEEK 7 - PYTHON - Pandas; Matplotlib; projects\21 FEB - Udemy project; London housing project\5. London Housing Data.csv")
udf2
```

# 1.  Convert the DATA TYPE of 'date' 

```python
udf2["date"] = udf2["date"].astype({'date': 'datetime64[ns]'})
udf2.info()
```

# 2. Add a new coloumn 'YEAR' in the dataframe, whihc controls years only

```python
udf2['year'] = pd.DatetimeIndex(udf2["date"]).year
udf2
```

# 2b. Add 'MONTH' as second coloumn, whihc controls months only

```python
udf2['month'] = pd.DatetimeIndex(udf2["date"]).month
udf2
```

# 3. Remove the coloumns 'YEAR' + 'MONTH' from the dataframe

```python
udf2 = udf2.drop(columns=['year', 'month'])
udf2
```

# 4. Show all records where 'No. of Crimes' is ZERO(0) + And how many records are there

```python
udf2[udf2["no_of_crimes"] == 0]
```

# 5. Mximum + minimum 'average_price' per year in England?

```python
udf2.groupby("area").agg({"average_price":["max","min"]})
```

# 6. Maximum + minimum No. of crimes recorded per area?

```python
udf2.groupby("area").agg({"no_of_crimes":["max","min"]})
```

# 7. Show the total count of records of each area, where average privce is less than 100,000

```python
udf2 = udf2[udf2['average_price'] < 100000]
udf2.groupby('area').agg({"average_price":["count"]})
```
